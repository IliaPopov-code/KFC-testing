## Here we construct a train and cal sets

```python
import os
import time
import numpy as np
import pandas as pd
```

```python
def f_downsample(data, n = 4):
    remainder = len(data) % n
    if remainder != 0:
        data = data[:len(data) - remainder]
    data_reshaped = np.array(data).reshape(-1, n)
    data_avg = data_reshaped.mean(axis=1)
    return data_avg

def get_strain_data(timecode, window, constant = 4096, detector = 'H1', downsample = True):
    if window >= 2:
        strain = TimeSeries.fetch_open_data(detector, timecode-window/2, timecode+window/2)
        white_data = strain.whiten()
        if downsample:
            bp_data = f_downsample(white_data.bandpass(30, 400)) #the constants that LIGO scientists use
        else:
            bp_data = white_data.bandpass(30, 400)
    else:
        strain = TimeSeries.fetch_open_data(detector, timecode-1, timecode+1)
        white_data = strain.whiten()
        bound = int(constant - 0.5*constant*window)
        shrunk_data = white_data.bandpass(30, 400)[bound: -bound]
        if downsample:
            bp_data = f_downsample(shrunk_data)
        else:
            bp_data = shrunk_data
    return bp_data
```

```python
window = 0.5         # Size of the window in seconds
constant = 4096      # Points per second (if needed elsewhere)
nulls = 50           # How many nulls per signal

# Global file path (update to your desired location)
global_file = '/Users/iliapopov/Desktop/e-BH/ebhconformal/signals_and_nulls.npz'

# Read all signal timecodes from CSV (assumes a column 'GPS')
signal_timecodes = np.array(pd.read_csv('/Users/iliapopov/Desktop/e-BH/ebhconformal/nonnulls.csv')['GPS'])

# --- Check if we've already saved data to resume from ---
if os.path.exists(global_file):
    loaded = np.load(global_file, allow_pickle=True)
    # Convert arrays stored as objects back to lists
    signal_set = loaded['signal_set'].tolist()
    null_set = loaded['null_set'].tolist()
    last_idx = int(loaded['last_idx'])
    print(f"Resuming from signal index {last_idx+1}")
    # Only process signals that haven't been processed yet.
    remaining_signals = signal_timecodes[last_idx:]
else:
    signal_set = []
    null_set = []
    last_idx = 0
    remaining_signals = signal_timecodes

# --- Processing loop ---
# Note: We use enumerate starting at last_idx+1 so that idx reflects the actual position in the original list.
for idx, signal in enumerate(remaining_signals, start=last_idx+1):
    # Process the signal data
    try:
        data = get_strain_data(signal, window)
        signal_set.append(data)
    
        # Process null iterations for this signal
        for null_idx in range(nulls):
            start_null_time = time.time()
            null_signal = signal + null_idx * 20 + 60
            null_data = get_strain_data(null_signal, window)
            null_set.append(null_data)
            elapsed = time.time() - start_null_time
            print(f"Processed null {null_idx+1}/{nulls} for signal {idx}/{len(signal_timecodes)} in {elapsed:.2f} seconds")

            # Update global file every 20 null iterations
            if (null_idx + 1) % 20 == 0:
                np.savez(global_file,
                         signal_set=np.array(signal_set, dtype=object),
                         null_set=np.array(null_set, dtype=object),
                         last_idx=idx)
                print(f"Intermediate global file update at signal {idx}, null iteration {null_idx+1}, time: {time.ctime()}")

        print(f"Completed signal {idx}/{len(signal_timecodes)}")
        # After finishing a signal, update the global file with the current index.
        np.savez(global_file,
                 signal_set=np.array(signal_set, dtype=object),
                 null_set=np.array(null_set, dtype=object),
                 last_idx=idx)
        print(f"Global file updated after completing signal {idx}, time: {time.ctime()}")
    except:
        pass
```

## Visualization of the sample creation

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from gwpy.timeseries import TimeSeries

# --- Define a simple downsample function ---
def downsample(data, n=4):
    """
    Downsamples the input array by averaging every n points.
    """
    data = np.asarray(data)
    # Truncate data so its length is a multiple of n
    remainder = len(data) % n
    if remainder != 0:
        data = data[:len(data) - remainder]
    # Reshape and average
    data_reshaped = data.reshape(-1, n)
    data_avg = data_reshaped.mean(axis=1)
    return data_avg

# --- Example parameters (update these as needed) ---
detector = 'H1'         # Example detector name
center = 1126259462.4      # Example GPS center time
window = 0.5              # Window parameter (seconds)
constant = 4096           # Example constant (used for cropping)


strain = TimeSeries.fetch_open_data(detector, center-1, center+1)
white_data = strain.whiten()

# --- Bandpass filter and crop (shrink) the data ---
bound = int(constant - 0.5 * constant * window)
shrunk_data = white_data.bandpass(30, 400)[bound: -bound]

# --- Downsample the cropped data by averaging every 5 points ---
bp_data = downsample(shrunk_data.value, n=5)

# --- Create time arrays for plotting ---
# The original time array (from strain)
t_orig = strain.times.value
# Whitened data time array
t_white = white_data.times.value
# Shrunk data time array
t_shrunk = shrunk_data.times.value
# For downsampled data, we create a time array spanning the shrunk data time interval
t_down = np.linspace(t_shrunk[0], t_shrunk[-1], len(bp_data))

# --- Plot each step ---
fig, axes = plt.subplots(4, 1, figsize=(12, 16), sharex=True)

# Plot 1: Original strain data
axes[0].plot(t_orig, strain.value, label="Original Strain", color='blue')
axes[0].set_title("Original Strain Data")
axes[0].legend()

# Plot 2: Whitened data
axes[1].plot(t_white, white_data.value, label="Whitened Data", color='orange')
axes[1].set_title("Whitened Data")
axes[1].legend()

# Plot 3: Bandpassed & Cropped (Shrunk) Data
axes[2].plot(t_shrunk, shrunk_data.value, label="Bandpassed & Cropped Data", color='green')
axes[2].set_title("Bandpassed & Cropped Data")
axes[2].legend()

# Plot 4: Downsampled Data
axes[3].plot(t_down, bp_data, label="Downsampled Data", color='red')
axes[3].set_title("Downsampled Data (Averaged Every 4 Points)")
axes[3].legend()

axes[3].set_xlabel("Time (s)")
plt.xlim(center-0.5, center+0.5)
plt.tight_layout()
plt.show()
```
